```python
source = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/horse-colic.csv"

columns = ['surgery','Age','Hospital ID','rectal temperature','pulse',
           'respiratory rate','temperature of extremities','peripheral pulse',
           "mucous membranes",'capillary refill time','pain','peristalsis',
           'abdominal distension','nasogastric tube','nasogastric reflux',
           'nasogastric reflux PH','rectal examination - feces','abdomen',
           'packed cell volume','total protein','abdominocentesis appearance',
           'abdomcentesis total protein','outcome','surgical lesion?','1','2','3',
           'path']

hc_df = pd.read_csv(source, names = columns)

hc_df.head().T
```

```python
# this variable is of no significance since pathology data is not included or collected for these cases
hc_df.drop(['path', 'Hospital ID'], axis=1, inplace=True)
hc_df.head().T
```

```python
hc_df.dtypes
```

## Replace '?' with np.nan

```python
for col in hc_df.select_dtypes(include='object'): # col is the column label
    hc_df[col] = hc_df[col].replace('?', np.nan)
hc_df.info()
```

From the features description, we infer the type of the variables i.e. categorical vs numerical. We then convert them to the appropriate dtypes.

## convert dtypes

```python
numerical = ['rectal temperature', 'pulse', 'respiratory rate', 'capillary refill time', 'packed cell volume', 'total protein']


for col in hc_df:
    if col in numerical:
        hc_df[col] = hc_df[col].astype('float64')
    else:
        hc_df[col] = hc_df[col].astype('category')

    
```

## Multiple variables in a single column

```python
lesion = hc_df['1']
regex = r'(1{2}|0{2}|[1-9]{1})([1,2,3,4]{1})([0,1,2]{1})(10|[0-9]{1})'
lesion = lesion.astype('str')
lesion_groups = lesion.str.extract(regex)
lesion_groups.info()
lesion_groups
```

```python
# add these columns back to the dataframe
hc_df['lesion site'] = lesion_groups[0].astype('category')
hc_df['lesion type'] = lesion_groups[1].astype('category')
hc_df['lesion subtype'] = lesion_groups[2].astype('category')
hc_df['lesion code'] = lesion_groups[3].astype('category')
```

```python
# Check how many values are non null
print((hc_df['2']==0).sum())
print((hc_df['3']==0).sum())

# since most of them are null, let's drop these columns
hc_df.drop(['2', '3'], axis=1, inplace=True)
```

## variables in rows 

```python
lesion_site = hc_df['lesion site'].replace({'1': 'gastric',
                              '2':'intestine',
                              '3': 'lg colon',
                              '4': 'lg colon and cecum',
                              '5' : 'cecum',
                              '6' : 'transverse colon',
                              '7' : 'retum/descending colon',
                             '8': 'uterus',
                              '9' : 'bladder',
                              '11' : 'all intestinal sites',
                             }
                            )
```

```python
# concatenate the dummies with the original dataframe
hc_df_2 = pd.concat([hc_df, pd.get_dummies(lesion_site, prefix='site')], axis=1)

hc_df_2.head().T
```

```python
hc_df_2 = pd.concat([hc_df_2, pd.get_dummies(lesion_type, prefix='type')], axis=1)

hc_df_2.head().T
```

```python
from statistics import variance
for col in ['rectal temperature', 'pulse', 'respiratory rate', ]:
    hc_df_2[col] = (hc_df_2[col] - hc_df_2[col].mean())/variance(hc_df_2[col])
    
hc_df_2.head().T
```

```python
hc_df_2.head().T
```

## Imputation by mean

```python
for col in hc_df_2.select_dtypes('float64'):
    print(col)
    print(hc_df_2[col].median())
    hc_df_2[col] =  hc_df_2[col].fillna(hc_df_2[col].median())
    print(hc_df_2[col].isna().sum())
```

## Imputation by mode

```python
for col in hc_df_2.select_dtypes('category'):
    print(col)
    print(hc_df_2[col].mode()[0])
    hc_df_2[col] =  hc_df_2[col].fillna(hc_df_2[col].mode()[0])
```
