## Goal 1 - Split bookname and chapter

Split the book name and chapter number into two columns: `Book` and `Chapter` using a string method. For example, `Buddhism_Ch1` becomes `Buddhism` and `Ch1`

```python
# split 'Unnamed: 0'
split = books_df['Unnamed: 0'].str.split('_', expand=True)

# Insert the 'Book' and 'Chapter' column in the first and second postion using DataFrame.insert() (Refer documentation)
books_df.insert(0, 'Chapter', split[1])
books_df.insert(0, 'Book', split[0])

# view the changes
books_df
```

```python
# drop the 'Unnamed: 0' column
books_df.drop('Unnamed: 0', axis=1, inplace=True)
books_df
```

## Goal 2 - Remove `Ch` from the chapter numbers

We are interested in removing `Ch` from the chapter numbers. Note that there is no space in between `Ch` and the number. 

```python
# remove the redundant `Ch` from the chapter numbers
books_df['Chapter'].replace('[Ch]', '', regex=True, inplace=True)

# view the changes
books_df
```
